```python
from alphablokus.train_utils import save_model_and_state, load_initial_state
from alphablokus.configs import NetworkConfig, GameConfig
```

```python
config_path = "../../configs/train_live/local_vast.toml"
learning_rate = 1e-3
training_file = "s3://alpha-blokus/full_v2/training_simulated/10m_10block_64chan_1768830688.pth"
new_file_name = "010260000"
new_model_directory = "s3://alpha-blokus/full_v2/models_v3/"
new_training_directory = "s3://alpha-blokus/full_v2/training_v3/"
```

```python
network_config = NetworkConfig(config_path)
game_config = GameConfig(config_path)
```

```python
model, optimizer = load_initial_state(
    network_config = network_config,
    game_config = game_config,
    learning_rate = learning_rate,
    device = "mps",
    training_file = training_file,
)
```

```python
save_model_and_state(
    model=model,
    optimizer=optimizer,
    name=new_file_name,
    model_directory=new_model_directory,
    training_directory=new_training_directory,
    device="mps",
)
```
